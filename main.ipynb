```python
# Step 1: Import required libraries
import boto3
import pandas as pd
import numpy as np
import sagemaker
from sagemaker import get_execution_role
from sagemaker.inputs import TrainingInput
from sagemaker.xgboost.estimator import XGBoost
import os
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from sagemaker.estimator import Estimator
import json

# Step 2: Load the data
train_data = pd.read_csv('train_data.csv')
test_data = pd.read_csv('test_data.csv')
print("Train Data Sample:")
print(train_data.head())
print("\nTest Data Sample:")
print(test_data.head())

# Step 3: Exploratory Data Analysis
print("\nTrain Data Shape:", train_data.shape)
print("Test Data Shape:", test_data.shape)
print("\nTrain Data Information:")
print(train_data.info())
print("\nDescriptive Statistics:")
print(train_data.describe())

# Check for class imbalance
print("\nTarget Distribution:")
print(train_data['cpu_state'].value_counts(normalize=True) * 100)

# Visualize data distributions
plt.figure(figsize=(12, 10))
sns.heatmap(train_data.corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Matrix')
plt.savefig('correlation_matrix.png')

# Step 4: Data Preprocessing
# Check for missing values
print("\nMissing Values in Train Data:")
print(train_data.isnull().sum())
print("\nMissing Values in Test Data:")
print(test_data.isnull().sum())

# Function to preprocess the data
def preprocess_data(data):
    # Create feature engineering
    # 1. Interaction terms
    data['temp_current'] = data['temperature'] * data['current']
    data['usage_voltage'] = data['cpu_usage'] * data['voltage']
    
    # 2. Squared terms for potentially non-linear relationships
    data['temp_squared'] = data['temperature'] ** 2
    data['usage_squared'] = data['cpu_usage'] ** 2
    
    # 3. Ratio features
    data['voltage_current_ratio'] = data['voltage'] / (data['current'] + 0.001)  # Avoid division by zero
    data['temp_fan_ratio'] = data['temperature'] / (data['fan_speed'] + 0.001)
    
    return data

# Apply preprocessing
train_data = preprocess_data(train_data)
test_data = preprocess_data(test_data)

# Split features and target
X_train = train_data.drop('cpu_state', axis=1)
y_train = train_data['cpu_state']
X_test = test_data.drop('cpu_state', axis=1)
y_test = test_data['cpu_state']

# Standardize numerical features
numerical_cols = ['temperature', 'voltage', 'current', 'cpu_usage', 'fan_speed', 
                  'temp_current', 'usage_voltage', 'temp_squared', 'usage_squared',
                  'voltage_current_ratio', 'temp_fan_ratio']
scaler = StandardScaler()
X_train[numerical_cols] = scaler.fit_transform(X_train[numerical_cols])
X_test[numerical_cols] = scaler.transform(X_test[numerical_cols])

# Step 5: Upload processed data to S3
bucket_name = 'predictive-maintanence'
prefix = 'cpu-state-xgboost-improved'
s3 = boto3.client('s3')
s3_resource = boto3.resource('s3')

# Prepare data in XGBoost format (label in first column)
train_data_xgb = pd.concat([y_train, X_train], axis=1)
test_data_xgb = pd.concat([y_test, X_test], axis=1)

train_file = 'train_xgb_improved.csv'
test_file = 'test_xgb_improved.csv'
validation_file = 'validation_xgb_improved.csv'

# Create validation set
X_train_final, X_val, y_train_final, y_val = train_test_split(
    X_train, y_train, test_size=0.2, random_state=42, stratify=y_train
)
val_data_xgb = pd.concat([y_val, X_val], axis=1)

# Save files locally
train_data_xgb.to_csv(train_file, header=False, index=False)
test_data_xgb.to_csv(test_file, header=False, index=False)
val_data_xgb.to_csv(validation_file, header=False, index=False)

# Upload to S3
train_s3_path = f's3://{bucket_name}/{prefix}/train/{train_file}'
test_s3_path = f's3://{bucket_name}/{prefix}/test/{test_file}'
validation_s3_path = f's3://{bucket_name}/{prefix}/validation/{validation_file}'

boto3.Session().resource('s3').Bucket(bucket_name).Object(f'{prefix}/train/{train_file}').upload_file(train_file)
boto3.Session().resource('s3').Bucket(bucket_name).Object(f'{prefix}/test/{test_file}').upload_file(test_file)
boto3.Session().resource('s3').Bucket(bucket_name).Object(f'{prefix}/validation/{validation_file}').upload_file(validation_file)

# Step 6: Train XGBoost model with improved hyperparameters
role = get_execution_role()

# Use the native SageMaker XGBoost estimator instead of the generic container
# Get the XGBoost container image URI
from sagemaker.image_uris import retrieve
xgboost_container = retrieve("xgboost", boto3.Session().region_name, version="1.5-1")


# Create the estimator
xgb = Estimator(
    image_uri=xgboost_container,
    role=role,
    instance_count=1,
    instance_type='ml.m5.large',
    output_path=f's3://{bucket_name}/{prefix}/output',
    sagemaker_session=sagemaker.Session()
)

# Set improved XGBoost hyperparameters
xgb.set_hyperparameters(
    max_depth=6,
    eta=0.1,  # Same as learning_rate
    gamma=0.1,
    min_child_weight=2,
    subsample=0.8,
    colsample_bytree=0.8,
    reg_alpha=0.01,
    reg_lambda=1,
    objective='multi:softprob',
    num_class=3,
    num_round=100,
    # Fix the evaluation metric - use a single value instead of a list
    eval_metric='mlogloss',  # You can use 'merror' here too, but pick one
    early_stopping_rounds=10,
    scale_pos_weight=1.0
)
# Step 7: Fit the model with validation data for early stopping
train_input = TrainingInput(train_s3_path, content_type='csv')
validation_input = TrainingInput(validation_s3_path, content_type='csv')

xgb.fit({
    'train': train_input,
    'validation': validation_input
})

# Step 8: Deploy the model to an endpoint
import time
endpoint_name = f'cpu-state-xgboost-endpoint-improved-{int(time.time())}'
xgb_predictor = xgb.deploy(
    initial_instance_count=1,
    instance_type='ml.m5.large',
    endpoint_name=endpoint_name
)

# Configure the serializer and deserializer for the predictor
xgb_predictor.serializer = sagemaker.serializers.CSVSerializer()
xgb_predictor.deserializer = sagemaker.deserializers.JSONDeserializer()

# Step 9: Enhanced evaluation with confusion matrix and detailed metrics
def evaluate_model(predictor, features, actual_labels, class_names=None):
    # Make predictions
    predictions = []
    prediction_probabilities = []
    
    # Process in batches to avoid memory issues
    batch_size = 100
    for i in range(0, len(features), batch_size):
        batch = features.iloc[i:i+batch_size].values.astype('float32')
        batch_csv = '\n'.join([','.join(map(str, row)) for row in batch])
        
        result = predictor.predict(batch_csv)
        
        # Extract predictions from result
        for j in range(len(result['predictions'])):
            scores = result['predictions'][j]['score']
            predicted_class = np.argmax(scores)
            predictions.append(predicted_class)
            prediction_probabilities.append(scores)
    
    predictions = np.array(predictions)
    
    # Calculate accuracy
    accuracy = np.mean(predictions == actual_labels)
    print(f"Test Accuracy: {accuracy * 100:.2f}%")
    
    # Calculate confusion matrix
    from sklearn.metrics import confusion_matrix, classification_report
    cm = confusion_matrix(actual_labels, predictions)
    
    # Plot confusion matrix
    plt.figure(figsize=(10, 8))
    if class_names is None:
        class_names = [f'Class {i}' for i in range(len(np.unique(actual_labels)))]
    
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title('Confusion Matrix')
    plt.savefig('confusion_matrix.png')
    
    # Detailed classification report
    print("\nClassification Report:")
    print(classification_report(actual_labels, predictions, target_names=class_names))
    
    return accuracy, predictions, prediction_probabilities

# Prepare test data for evaluation
test_features = X_test
actual_labels = y_test.values

# Class names for better visualization
class_names = ['Normal', 'Warning', 'Critical']  # Update with your actual class names

# Evaluate the model
accuracy, predictions, probabilities = evaluate_model(xgb_predictor, test_features, actual_labels, class_names)

# Step 10: Feature importance analysis
# Get feature importances from the model
try:
    model_artifacts = sagemaker.s3.S3Downloader.list(xgb.model_data)
    model_path = sagemaker.s3.S3Downloader.download(xgb.model_data)
    
    # If we can access feature importance, let's plot it
    import xgboost as xgb
    model = xgb.Booster()
    model.load_model(model_path)
    
    importance_type = 'weight'  # or 'gain', 'cover'
    feature_importance = model.get_score(importance_type=importance_type)
    
    # Sort features by importance
    sorted_importance = sorted(feature_importance.items(), key=lambda x: x[1], reverse=True)
    features, importance = zip(*sorted_importance)
    
    plt.figure(figsize=(12, 8))
    sns.barplot(x=list(importance), y=list(features))
    plt.title(f'Feature Importance ({importance_type})')
    plt.tight_layout()
    plt.savefig('feature_importance.png')
    
    print("\nTop 10 Important Features:")
    for i, (feature, imp) in enumerate(sorted_importance[:10]):
        print(f"{i+1}. {feature}: {imp}")
except Exception as e:
    print(f"Couldn't access feature importance: {e}")

# Step 11: Clean up (uncomment when ready to delete resources)
# print(f"Deleting endpoint {endpoint_name}")
# sagemaker.Session().delete_endpoint(endpoint_name)
```
